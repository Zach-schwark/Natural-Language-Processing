Data set from : https://huggingface.co/datasets/masakhane/masakhanews

File to preprocess the news dataset to get the lables we need

Original Labels:

label_id, label\
0:       business\
1:      entertainment\
2:      health\
3:      politics\
4:      religion\
5:      sports\
6:      technology



```python
import pandas as pd
from datasets import load_dataset

data = load_dataset('masakhane/masakhanews', 'swa') 

display(data['train'][0])
```

```python
def change_label(data_point):
    if data_point["label"] == 3:
        data_point["label"] = 1
    else:
        data_point["label"] = 0
    return data_point

politics_label_data = data.map(change_label)

politics_label_data["train"][:5]
```
